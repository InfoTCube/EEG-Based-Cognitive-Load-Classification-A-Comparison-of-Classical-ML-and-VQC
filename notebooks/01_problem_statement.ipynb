# **EEG-Based Cognitive Load Classification: A Comparison of Classical Machine Learning and Variational Quantum Classifiers**

Cognitive load detection from EEG is critical for occupational safety and adaptive learning systems. 
We develop and compare classical ML (Random Forest, XGBoost, SVM) with quantum approaches (VQC) 
for classifying mental math vs. resting states.

**Impact:** Establishes realistic performance baseline and evaluates quantum advantage in biomedical ML.

## Background

Cognitive overload affects 15-20% of drivers annually, contributing to 1M accidents/year globally.
Current detection methods:
- Subjective questionnaires (unreliable, post-hoc)
- Eye-tracking (expensive, requires special equipment)
- Manual EEG analysis (time-consuming, requires expertise)

Automated EEG classification could enable real-time detection in vehicles, classrooms, operating rooms.

EEG captures real-time brain activity via frequency bands:
- Theta (4-8 Hz): Increases with working memory load
- Beta (12-30 Hz): Increases during active thinking
- Alpha (8-12 Hz): Decreases with attention demands

Unlike fMRI (expensive, immobile) or eye-tracking (limited to visual tasks), 
EEG enables portable, scalable cognitive monitoring.

#### Why Machine Learning?
- Manual analysis is not scalable
- ML can learn patterns humans can't see
- Enables automation and real-time decisions

#### Why Compare Classical + Quantum?
- Classical ML is baseline (proven, fast)
- Quantum ML is emerging (potential advantage, untested)
- Rigorous comparison is needed for technology evaluation

## Dataset Overview

#### Experimental Protocol
Subjects: 36 healthy adults (18-40 years)

Conditions:
1. Rest: 60 sec eyes open (baseline)
2. Mental Math: 60 sec serial subtraction by 7 (cognitive load task)

Equipment:
- 21-channel EEG (10-20 system)
- 500 Hz sampling rate
- Bandpass filter: 1-50 Hz

Total: 120 sec/subject × 21 channels × 500 Hz = 1.26M data points per subject

## Research questions

Can we reliably classify cognitive load states (normal vs. mental math) from EEG with cross-subject generalization using quantum machine learning models?

**Secondary Questions:**
1. Which EEG features best discriminate cognitive load?
2. Do classical algorithms (RF, XGBoost, SVM) differ significantly?
3. Does quantum ML provide any advantage over classical?
4. What is the realistic performance ceiling?
